
Boxfiltering , Median Gussian Filtering




```python

import matplotlib.pyplot as plt 
import cv2
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img4 = cv2.imread('/content/me_4.jpg')
blur = cv.blur(img4, (100,100))
plt.subplot(121),plt.imshow(img4), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show
```

```python
# median filtering

blur = cv.medianBlur(img4, 105)
plt.subplot(121),plt.imshow(img4), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show
```

```python
# GaussianBlur

blur = cv.GaussianBlur(img4, (105,105), 0)

plt.subplot(121),plt.imshow(img4), plt.title('original')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(blur),plt.title('Blurred')
plt.xticks([]),plt.yticks([])
plt.show
```
