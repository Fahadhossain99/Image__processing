Subplot,histogram,equalization


```python
import matplotlib.pyplot as plt
import cv2
```

```python
I1 = cv2.imread("/content/me.jpg")
I2 = cv2.imread("/content/me_1.jpg")
plt.subplot(2,2,1), plt.imshow(I1), plt.title("image-1")
plt.subplot(2,2,2), plt.imshow(I2), plt.title("image-2")
```

```python
img = cv2.imread("/content/me_2.jpg")
plt.imshow(img)
```

```python
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.title("RGB Image")
```

```python
img2 = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img1)
plt.title("GRAY Image")
```

```python
plt.imshow(img2, cmap='gray')
```

```python
hist = cv2.calcHist(img2,[0],None,[256],[0,255])
plt.plot(hist)
plt.show()
```

```python
eq = cv2.equalizeHist(img2)
plt.plot(eq)
plt.show()
```
